```python
import pandas as pd
import json
import itertools
```

```python
pd.set_option('display.max_rows', 500)

with open("jobs.json") as f:
    json_file = json.load(f)

place_names = [
    "stourbridge", "kidderminster", "wolverhampton", "birmingham", "hagley", "kinver", "lye", "bridgnorth",
    "halesown", "smethwick", "worcester", "kingswinford", "dudley", "bromsgrove"
]

remote_names = [
    "home working", "remote working", "homeworking", "home-based", "home", "home based", "homebased", "remote",
    "remote based", "remote-based"
]

accepted_job_names = [
    "data", "developer", "software", "it", "web", "it/web", "analyst", "engineer", "cad"
]

accepted_place_names = place_names + remote_names


def get_jobs(file, date, job_names, location_names):
    return [
        {**i, "date": date} for i in file[date] 
        if i["location"].split(",")[0].lower() in location_names and
        any(True for j in i["title"].split(" ") if j.lower() in job_names)
    ]


s_table = pd.DataFrame(
    itertools.chain.from_iterable(
        [get_jobs(json_file, i, accepted_job_names, accepted_place_names) for i in json_file.keys()]
    )
)
s_table
```
